```python
def normalize(image):
    image = cv2.normalize(image, None, 0, 1, cv2.NORM_MINMAX)
    return image
```

```python
def preprocess_all_images(images, target_size=(256, 256)):
    processed_images = []
    for img_path in images:
        if isinstance(img_path, str):
            img = cv2.imread(img_path)
            if img is not None:
                img = cv2.resize(img, target_size)
                processed_images.append(img)
            else:
                print(f"Failed to load image: {img_path}")
        else:
            print(f"Image path is not a string: {img_path}")

    return processed_images
```

```python
def preprocess_images(images, target_size=(256, 256)):
    processed_images = []
    for img_path in images:
        print(f"Attempting to read image from path: {img_path}")
        if isinstance(img_path, str):
            img = cv2.imread(img_path)
            if img is not None:
                img = cv2.resize(img, target_size)
                processed_images.append(img)
            else:
                print(f"Failed to load image: {img_path}")
        else:
            print(f"Image path is not a string: {img_path}")

    return processed_images
```

```python
def combine_masks(masks):
    if not masks:
        return None

    combined_mask = masks[0].copy()
    for mask in masks[1:]:
        combined_mask = cv2.bitwise_or(combined_mask, mask)

    return combined_mask
```

```python
all_processed_masks = []

for i in range(len(mul_masks)):
    masks_for_case = preprocess_images(mul_masks[i])

    if masks_for_case:
        combined_mask = combine_masks(masks_for_case)
        all_processed_masks.append(combined_mask)
```

### Segmentation

```python
def apply_mask(image, mask):
    return image * mask
```

```python
segmented_images = []
for img, mask in zip(us_images, tumor_masks):
    segmented_img = apply_mask(img, mask)
    segmented_images.append(segmented_img)
```

### saving and downloading segmented images
